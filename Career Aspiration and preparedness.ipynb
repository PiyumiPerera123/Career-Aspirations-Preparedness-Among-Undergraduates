```python
def convert(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_float(x):
    if type(x) == float:
        return x
    else:
        x = x.strip()
        tockens = x.split(' ')
        return float(tockens[0])
```

```python
df2 = df1.copy()
df2['GPA_Credits'] = df2['GPA_Credits'].apply(convert_float)
```

```python
df2.dtypes
```

### Take a sample using poststratification

```python
df2['Degree_Type'].value_counts()
```

```python
research = df2[df2['Degree_Type']=='Research Oriented Honors degree']
industry = df2[df2['Degree_Type']=='Industry Oriented Honors degree']
general = df2[df2['Degree_Type']=='General degree']
```

```python
research_sample = research.sample(52,random_state=42)
industry_sample = industry.sample(35,random_state=42)
general_sample = general
```

```python
df2 = pd.concat([research_sample,industry_sample,general_sample],axis=0)
```

```python
df2.shape
```

```python
df2['Degree_Type'].value_counts()
```

```python
df2.to_csv('Career Aspiration and Preparedness sample.csv',index=False)
```

```python
def create_pie(column):
    count = df2[column].value_counts()
    size = count.values
    labels = count.index

    plt.pie(size,autopct='%1.1f%%',wedgeprops={'width':0.3},startangle=90)
    plt.legend(labels,loc='upper right',fontsize='xx-small',bbox_to_anchor=(2,0.8))
    plt.title(column)
    plt.show()
```
